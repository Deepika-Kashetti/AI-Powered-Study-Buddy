```python
import google.generativeai as genai

genai.configure(api_key="Enter your Gemini API Key")

model = genai.GenerativeModel("models/gemini-2.5-flash")

topic = "Artificial Intelligence"

# Explanation
print("========== EXPLANATION ==========\n")
explanation = model.generate_content(
    f"Explain {topic} in simple words for students."
)
print(explanation.text)

# Summary
print("\n========== SUMMARY ==========\n")
summary = model.generate_content(
    f"Summarize {topic} in 5 bullet points."
)
print(summary.text)

# Quiz
print("\n========== QUIZ QUESTIONS ==========\n")
quiz = model.generate_content(
    f"Generate 5 quiz questions on {topic}."
)
print(quiz.text)

# Flashcards
print("\n========== FLASHCARDS ==========\n")
flashcards = model.generate_content(
    f"Generate 5 flashcards for {topic}."
)
print(flashcards.text)
```
